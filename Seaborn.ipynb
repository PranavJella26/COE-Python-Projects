```python
# Insights
The dataset contains 244 records with no missing values.
The average total bill is about 19.79, while the average tip is around 3.00.
Most groups consist of 2–3 people, with an average group size of about 2.57.
The minimum total bill is 3.07 and the maximum is 50.81 show the spreading .
Tips range from 1.00 to 10.00, shows the variation in tip.
The median total bill (17.80) is slightly lower than the mean, suggesting some high-value bills.
The 75% of customers have group sizes of 3 or fewer, showing smaller groups are more common.
```
